#Multiple Seasons

```python
leagues = ['E0','E2', 'E3']

frames = []

# loop through range of seasons and insert new column
for league in leagues:
    for season in range(15,21): # range of seasons
        df = pd.read_csv(root + str(season) + str(season+1) + "/" + league + ".csv")
#         df['season'] = season
        df.insert(1, 'season', season)
        frames.append(df)
```

```python
# total frames (6 seasons x 3 leagues)
len(frames)
```

```python
# show 1st, and last element (new season column was created)
frames[17]
```

# Organize All The Data in a Dictionary

```python
# create a dictionary with original name of league as key
dict_countries = {
              'Spanish La Liga':'SP1', 'Spanish Segunda Division':'SP2',
              'German Bundesliga':'D1',
              'English Premier League':'E0', 'English League 1':'E2', 'English League 2':'E3',
             }
```

```python
# get a dictionary element
dict_countries['German Bundesliga']
```

```python
dict_historical_data = {}

# looping through key elements
for league in dict_countries:
    frames = [] # one dataframe per league
    for season in range(15, 21):
        df = pd.read_csv(root + str(season) + str(season+1) + "/" + dict_countries[league]+ ".csv")
        df.insert(1, 'season', season)
        frames.append(df)
    df_frames = pd.concat(frames) # storing all seasons of a league in a single dataframe
    dict_historical_data[league] = df_frames # assigning dataframe to a new dictionary
```

```python
# show dataframe inside dict_historical_data
dict_historical_data['Spanish La Liga']
```

# Creating a conditional colun from more than 2 choices: np.select()

```python
import numpy as np
```

```python
# Create an array based on multiple price tiers (+2 choice)

# Create conditions and values variables
conditions = [
    (df_laptops['Price_euros']>3000),
    (df_laptops['Price_euros']>2000) & (df_laptops['Price_euros']<=3000),
    (df_laptops['Price_euros']>800) & (df_laptops['Price_euros']<=2000),
    (df_laptops['Price_euros']<=800),
]
choices = ['Too Expensive', 'Expensive', 'Affordable', 'Cheap']
```

```python
# set it to a new column
df_laptops['Price_tier'] = np.select(conditions, choices, default='Unknown')
```

```python
# show dataframe
df_laptops.head(5)
```

```python
# Count values inside price_tier column
df_laptops['Price_tier'].value_counts()
```

```python
# create an array based on multiple screen size tiers (+2 choices)
# Too Big>16, Big>14, Small>12, Too Small<12

# create "conditions" and "values" variables
conditions = [
    (df_laptops['Inches']>16),
    (df_laptops['Inches']>14) & (df_laptops['Inches']<=16),
    (df_laptops['Inches']>12) & (df_laptops['Inches']<=14),
    (df_laptops['Inches']<=12),
]
choices = ['Too Big', 'Big', 'Small', 'Too Small']
```

```python
# set it to a new column
df_laptops['Screen_size'] = np.select(conditions, choices, default='Unknown')
```
